```python
from sklearn.tree import DecisionTreeClassifier
```

```python
features = [
    [150, 0],  # Apple
    [170, 0],  # Apple
    [130, 1],  # Orange
    [120, 1],  # Orange
]
```

```python
labels = ["apple", "apple", "orange", "orange"]
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
prediction = clf.predict([[130,0]])
```

```python
print(f"The fruit is {prediction[0]}")
```
