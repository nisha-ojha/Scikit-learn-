```python
import pandas as pd
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_excel("iris-train.xlsx")
```

```python
data.head()
```

```python
x = data.iloc[:,:-1]  #take all the rows and all the columns except the last one
```

```python
x.values
```

```python
y = data.iloc[:, -1]
```

```python
y.values
```

```python
model= GaussianNB()
```

```python
model.fit(x,y)
```

```python
predictions = model.predict([[5.4, 3,4.5, 1.5]])
```

```python
predictions
```

```python
datatest = pd.read_excel("iris.xlsx")
```

```python
x_test = datatest.iloc[:,:-1]
```

```python
predictions = model.predict(x_test.values)
```

```python
predictions
```

```python
y_test = datatest.iloc[:,-1]
```

```python
actual = y_test.values
```

```python
count = 0
for i in range(0, len(actual)):
    if actual[i] == predictions[i]:
        count = count + 1
print((count*100)/len(actual))
```
